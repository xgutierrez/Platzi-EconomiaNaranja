## 2. EDA: Exploratory Data Analysis

Consiste en visualizar los datos antes de las fórmulas estadísticas.


### Scatterplot

Características:

* Permite cruzar variables continuas o numéricas, nuca textos
* Se diferencia a una gráfica de líneas porque los puntos generados no los podemos unir.
* Cada punto es una coordenada que no se pueden unir
* La variable independiente se pone en el eje X y la variable dependiente en el eje Y

```python
import matplotlib.pyplot as plt
```

#### Gráficas del dataset mtcars

```python
# Scatter Plot

x = mtcars['cyl']
y = mtcars['mpg']

plt.scatter(x,y)
plt.title('Relación cilindros y millas por galón')
plt.xlabel('cilindros')
plt.ylabel('millas por galón')
plt.show()
```

```python
# Scatter Plot

x = mtcars['hp']
y = mtcars['mpg']

plt.scatter(x,y)
plt.title('Relación caballos de fuerza y millas por galón')
plt.xlabel('caballos de fuerza')
plt.ylabel('millas por galón')
plt.show()
```

```python
# Scater plot con seaborn

import seaborn as sns

ax = sns.scatterplot(x="hp",y="mpg", data=mtcars)
```

#### Gráficas del dataset de Economía Naranja

```python
# Scatter Plot

x = orangeec['Education invest % GDP']
y = orangeec['Unemployment']

plt.scatter(x,y)
plt.title('Relación inversión en educación y desempleo')
plt.xlabel('Inversión educación (%PIB)')
plt.ylabel('Desempleo')
plt.show()
```

```python
# Scatter Plot

x = orangeec['AporteEcNja']
y = orangeec['GDP PC']

plt.scatter(x,y)
plt.title('Relación economía naranja y PIB per capital')
plt.xlabel('Aporte economía naranja al PIB(%)')
plt.ylabel('PIB per cápital')
plt.show()
```

### Histogramas

Características:

* Sirve para ver la distribución de una variable 
* Las barras van pegadas
* No se reorganiza el eje X y va de mayor a menor
* El histograma me presenta lo que hay y no hay (Por ejemplo en la gráfica me indica que no hay frecuencias para edades de 45 y 47)
* Siempre se tendrá variables continuas y númericas

```python
x = mtcars['hp']

plt.hist(x,normed=False,bins = 30)
plt.ylabel('Números de Carros')
plt.xlabel('Caballos de Fuerza')
plt.title('Carros según caballos de fuerza')
plt.grid(True)
plt.show()
```

### BoxPlot

```python
#Box Plot en Matplotlib
#Datos
x = mtcars['hp']
y =mtcars['cyl']
#Definición para los outliers
red_square = dict(markerfacecolor='r',marker='s')

#Gráfico del boxplot
plt.boxplot((x,y),flierprops=red_square)

#Para poner etiqueta en el eje y
plt.ylabel('Caballos de Fuerza')
#Para poner título a la gráfica
plt.title('Caballos de fuerza en carros mtcars')

#Mostrar el gráfico
plt.show()
```

```python
# Boxplot en pandas

mtcars.boxplot(column=['hp'],by=['cyl'])
```

```python
#Boxplot con Seaborn
#Se crea una BoxPlot haciendo que una variable dependa de la otra

import seaborn as sns

y = mtcars['hp']
x = mtcars['cyl']

sns.set(style="whitegrid")

#ax = sns.boxplot(x)

ax = sns.boxplot(x,y)
```
